```python
import pandas as pd
import numpy as np
import os.path as osp
import os
import re
from collections import OrderedDict
```

```python
BASE_PATH = osp.join("..", "slurm_logs")
LOG_FNAMES = [
 'cogs_inv_ppl_htl_10_0.33-5636179.log',
 'cogs_chia_htl_10_0.33-5636184.log',
 'cogs_full_42-5636185.log'
]
```

```python
COGS_LINE_REGEX = r"In:\t.*'(.*)'.*Out:\t(.*)\tType:\t(.*)\tRef:\t(.*)"
```

```python
MISTAKE_EXAMPLES = [
    OrderedDict(),
    OrderedDict(),
    OrderedDict()
]

for IDX, LOG_FNAME in enumerate(LOG_FNAMES):
    CUR_PATH = osp.join(BASE_PATH, LOG_FNAME)
    with open(CUR_PATH, "r") as f:
        for line in f:
            result = re.search(COGS_LINE_REGEX, line)
            if result:
                MISTAKE_EXAMPLES[IDX][result.groups()] = True
```

```python
MISTAKE_EXAMPLES = [set(MISTAKE_EXAMPLE.keys()) for MISTAKE_EXAMPLE in MISTAKE_EXAMPLES]
```
